```python
from np_fingerprint_benchmark.data_integration.get_info_from_plantcyc import data_by_record

pathways = data_by_record("pathways.dat")
reactions = data_by_record("reactions.dat")
```

```python
pathway_reactions = {}
for pathway in pathways:
    pathway_reactions[pathway["UNIQUE-ID"][0]] = pathway["REACTION-LIST"]

reactions_ec = {}
for reaction in reactions:
    if "EC-NUMBER" in reaction:
        ec_numbers = [ec.replace("EC-", "") for ec in reaction["EC-NUMBER"]]
        reactions_ec[reaction["UNIQUE-ID"][0]] = ec_numbers
```

```python
import pickle

with open('pathway_reactions.pkl', 'wb') as f:
    pickle.dump(pathway_reactions, f)

with open('reactions_ec.pkl', 'wb') as f:
    pickle.dump(reactions_ec, f)
```
