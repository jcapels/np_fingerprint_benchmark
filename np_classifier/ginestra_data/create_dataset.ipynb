```python
import pandas as pd

merged_dataset = pd.read_csv("../merged_dataset.csv")
```

```python
import pandas as pd

train_ginestra = pd.read_csv("train_smiles.csv")
validation_ginestra = pd.read_csv("val_smiles.csv")
test_ginestra = pd.read_csv("test_smiles.csv")

train = merged_dataset[merged_dataset["SMILES"].isin(train_ginestra["SMILES"])]
validation = merged_dataset[merged_dataset["SMILES"].isin(validation_ginestra["SMILES"])]
test = merged_dataset[merged_dataset["SMILES"].isin(test_ginestra["SMILES"])]
```

```python
train.to_csv("train_ginestra.csv", index=False)
validation.to_csv("val_ginestra.csv", index=False)
test.to_csv("test_ginestra.csv", index=False)
```
