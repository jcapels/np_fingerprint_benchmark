```python
import pandas as pd
```

```python
train_dataset = pd.read_csv("train_dataset.csv")
validation_dataset = pd.read_csv("validation_dataset.csv")
```

```python
merged_dataset = pd.concat((train_dataset, validation_dataset))
merged_dataset = merged_dataset.sample(frac=1)
```

```python
from skmultilearn.model_selection import iterative_train_test_split
from skmultilearn.model_selection import IterativeStratification

k_fold = IterativeStratification(n_splits=5, order=1)
train_idx = []
test_idx = []
for train, test in k_fold.split(merged_dataset.key, merged_dataset.iloc[:, 2:]):
    train_idx.append(train)
    test_idx.append(test)
```

```python
merged_dataset = merged_dataset.reset_index(drop=True)
```

```python
merged_dataset.to_csv("merged_dataset.csv", index=False)
```

```python
import pickle

with open("train_idx.pkl", "wb") as f:
    pickle.dump(train_idx, f)

with open("test_idx.pkl", "wb") as f:
    pickle.dump(test_idx, f)
```
