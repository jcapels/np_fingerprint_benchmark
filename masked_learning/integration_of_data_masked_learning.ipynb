# Split the data

```python
from deepmol.loaders import CSVLoader

dataset = CSVLoader(dataset_path="integrated_dataset_km_training_wo_fintuning_datasets.csv", smiles_field="smiles", id_field="ids").create_dataset(sep=",")
```

```python
from deepmol.splitters import ScaffoldSplitter

scaffold_splitter = ScaffoldSplitter()

train_dataset, valid_dataset, test_dataset = scaffold_splitter.train_valid_test_split(dataset, frac_train=0.8, frac_valid=0.1, frac_test=0.1, homogenous_datasets = False)
```

```python

train_dataset.to_csv(path="train_test_split_scaffolds/train_dataset.csv", index=False)
valid_dataset.to_csv(path="train_test_split_scaffolds/validation_dataset.csv", index=False)
test_dataset.to_csv(path="train_test_split_scaffolds/test_dataset.csv", index=False)
```

```python
from deepmol.loaders import CSVLoader

loader = CSVLoader(dataset_path="train_test_split_scaffolds/train_dataset.csv",
                           smiles_field='smiles',
                           id_field='ids',
                           mode='auto')
train_dataset = loader.create_dataset(sep=',')

loader = CSVLoader(dataset_path="train_test_split_scaffolds/validation_dataset.csv",
                           smiles_field='smiles',
                           id_field='ids',
                           mode='auto')
validation_dataset = loader.create_dataset(sep=',')

loader = CSVLoader(dataset_path="train_test_split_scaffolds/test_dataset.csv",
                           smiles_field='smiles',
                           id_field='ids',
                           mode='auto')
test_dataset = loader.create_dataset(sep=',')
```

```python
from rdkit import Chem
from rdkit.Chem import AllChem, DataStructs
from tqdm import tqdm

def compute_fingerprint(molecule):
    fingerprint = AllChem.GetMorganFingerprintAsBitVect(molecule, 2, nBits=1024)
    return fingerprint

# Compute molecular fingerprints for each molecule in the combined dataset
train_dataset_fingerprints = [compute_fingerprint(smiles) for smiles in train_dataset.mols]
test_dataset_fingerprints = [compute_fingerprint(smiles) for smiles in test_dataset.mols]
```

```python
validation_dataset_fingerprints = [compute_fingerprint(smiles) for smiles in validation_dataset.mols]
```

```python
max_score = 0
more_than_0_6 = 0

to_drop = []

import numpy as np
from tqdm import tqdm

for i in tqdm(range(len(validation_dataset_fingerprints))):
    scores = DataStructs.BulkTanimotoSimilarity(validation_dataset_fingerprints[i], train_dataset_fingerprints)
    max_score_ = np.max(scores)
    if max_score < max_score_:
        max_score = max_score_
    
    if max_score_ > 0.6:
        to_drop.append(i)
```

```python
validation_dataset.remove_elements_by_index(to_drop, inplace=True)
```

```python
validation_dataset.to_dataframe().to_csv("train_test_split_scaffolds/validation_dataset_wo_redundancy.csv")
```

```python
max_score = 0
more_than_0_6 = 0

to_drop = []

import numpy as np
from tqdm import tqdm

for i in tqdm(range(len(test_dataset_fingerprints))):
    scores = DataStructs.BulkTanimotoSimilarity(test_dataset_fingerprints[i], train_dataset_fingerprints)
    max_score_ = np.max(scores)
    if max_score < max_score_:
        max_score = max_score_
    
    if max_score_ > 0.6:
        to_drop.append(i)

test_dataset.remove_elements_by_index(to_drop, inplace=True)
test_dataset.to_dataframe().to_csv("train_test_split_scaffolds/test_dataset_wo_redundancy.csv")
```
